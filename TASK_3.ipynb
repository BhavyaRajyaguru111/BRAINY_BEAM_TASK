```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Set up the figure and axis
fig, ax = plt.subplots()
xdata = np.linspace(0, 2 * np.pi, 100)
ydata = np.sin(xdata)
line, = ax.plot(xdata, ydata, color='blue')

# Set up the axis limits
ax.set_ylim(-1.5, 1.5)
ax.set_xlim(0, 2 * np.pi)

# Function to update the plot
def update(frame):
    line.set_ydata(np.sin(xdata + frame / 10))  # Update the y data
    return line,

# Create the animation
ani = FuncAnimation(fig, update, frames=np.arange(0, 100), blit=True)

# Show the plot
plt.title('Dynamic Line Chart Animation')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```
