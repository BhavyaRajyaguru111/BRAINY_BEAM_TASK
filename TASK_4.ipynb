```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a figure and a set of subplots
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the sine wave
ax.plot(x, y1, label='Sine Wave', color='blue', linestyle='-', linewidth=2)

# Plot the cosine wave
ax.plot(x, y2, label='Cosine Wave', color='red', linestyle='--', linewidth=2)

# Customize axes
ax.set_title('Sine and Cosine Waves', fontsize=16, fontweight='bold')
ax.set_xlabel('X-axis', fontsize=14)
ax.set_ylabel('Y-axis', fontsize=14)
ax.set_xlim(0, 10)
ax.set_ylim(-1.5, 1.5)

# Add grid
ax.grid(True, linestyle='--', alpha=0.7)

# Add legend
ax.legend(loc='upper right', fontsize=12)

plt.show()

```
